## BOXPLOT-
A box plot gives a five-number summary of a set of data which is-
1. Minimum - It is the minimum value in the dataset excluding the outliers.
2. First Quartile (Q1) - 25% of the data lies below the First (lower) Quartile.
3. Median (Q2) - It is the mid-point of the dataset. Half of the values lie below it and half above.
4. Third Quartile (Q3) - 75% of the data lies below the Third (Upper) Quartile.
5. Maximum - It is the maximum value in the dataset excluding the outliers.

## LINEAR REGRESSION-
1. Simple and effective for continuous variable prediction.
2. Identifies patterns in study habits affecting grades.
3. Helps make recommendations (e.g., optimizing study hours for better performance).

```python
#Libraries for Linear Regression
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, r2_score, mean_absolute_error
```

```python
# Select Features for Prediction
x = student[['study_hours_per_day', 'sleep_hours', 'screen_time']]
y = student['exam_score']  # Predicting grades based on habits
```

```python
# Train-Test Split
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=42)
```

```python
# Initialize
model = LinearRegression()
```

```python
# Train the Model
model.fit(x_train, y_train)
```

```python
# Predictions
y_pred = model.predict(x_test)
```

```python
# Model Evaluation
mse = mean_squared_error(y_test, y_pred)
rmse = np.sqrt(mean_squared_error(y_test,y_pred))
r2 = r2_score(y_test, y_pred)
```

```python
print("Mean Squared Error: ",mse)
print("Root Mean Squared Error: ",rmse)
print("R-Squared Score: ",r2)
```

```python
# Visualization
plt.scatter(y_test, y_pred, color="pink")
plt.xlabel("Actual Grades")
plt.ylabel("Predicted Grades")
plt.title("Actual vs Predicted Grades")
plt.grid()
plt.show()
```

Key Takeaways from Predictive Analysis-
1. Higher study hours generally correlate with better grades.
2. Limited screen time may improve students' academic performance.
3. Sleep patterns play a significant role in overall focus and efficiency.

## CLASSIFICATION MODELLING-
It is used to categorize students into different performance levels based on their habits. 
We'll use a Random Forest Classifier, a powerful and accurate model for classification tasks.

```python
#Import libraries for Classification
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix
```

```python
# Create Performance Categories (Low, Medium, High)
student['performance_level'] = pd.cut(student['exam_score'], bins=[0, 60, 80, 100], labels=['Low', 'Medium', 'High'])
```

```python
# Encode Categorical Labels
student['performance_level'] = student['performance_level'].astype('category').cat.codes  # Convert to numeric
```

```python
# Selecting Features for Classification
x = student[['study_hours_per_day', 'sleep_hours', 'screen_time']]
y = student['performance_level']  # Target: Performance level category
```

```python
# Splitting Data into Training and Testing Sets
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=42)
```

```python
# Initialize
model = RandomForestClassifier(n_estimators=100, random_state=42)
```

```python
#Train Model
model.fit(x_train, y_train)
```

```python
# Predictions
y_pred = model.predict(x_test)
```

```python
# Model Evaluation
accuracy = accuracy_score(y_test, y_pred)
print("Model Accuracy: ",accuracy)
```

```python
print("Classification Report:\n", classification_report(y_test, y_pred))
```

```python
# Confusion Matrix Visualization
sns.heatmap(confusion_matrix(y_test, y_pred), annot=True, fmt='d', cmap="Blues")
plt.xlabel("Predicted")
plt.ylabel("Actual")
plt.title("Confusion Matrix")
plt.show()
```

Key Insights from Classification Modelling-
1. Groups students into performance categories based on their habits.
2. Provides actionable insights for personalized study plans.
3. Helps educators identify at-risk students for intervention.
